```python
import json
from IPython.display import Markdown, display

def visualmind(input_usuario):
    """
    Simulación del motor de IA que optimiza consultas (Fast Prompting).
    """
    prompt_etapa_1 = "Actúa como un Arquitecto de Información. Genera una salida JSON..."
    prompt_etapa_2 = "Actúa como Diseñador Instruccional experto en Método Feynman..."

    # Respuesta estructurada (JSON)
    # Sobre Ciclo Nuclear (ejemplo)
    datos_extraidos = {
        "conceptos_clave": [
            {"término": "Ciclo del combustible nuclear", "definición": "Procesos industriales para extracción y uso de uranio."},
            {"término": "Enriquecimiento", "definición": "Incremento de la proporción de isótopos fisionables."}
        ],
        "punto_critico": "Diferencia funcional entre ciclo abierto y ciclo cerrado.",
        "analogia_potencial": "Es similar a un proceso de reciclaje doméstico."
    }

    # Metodo Feynman
    metodo_feynman = """
    ### Guion de Micro-Learning
    **1. Gancho:** ¿Sabías que el combustible nuclear puede usarse una sola vez... o reaprovecharse como reciclaje?
    **2. Explicación:** En un ciclo abierto, usas y tiras al vertedero. En un ciclo cerrado, separas lo reciclable (reprocesamiento).
    **3. Conclusión:** Entender esto ayuda a decidir cómo gestionamos lo que dejamos atrás.
    """

    # Prompt de Imagen (Texto-Imagen) en Inglés
    prompt_imagen = "Generate a high-quality educational infographic asset... Clean 3D Isometric Illustration..."

    return datos_extraidos, metodo_feynman, prompt_imagen

print("Motor de VisualMind configurado con éxito.")
```

```python
texto_a_procesar = "https://youtu.be/hJME9ACESKM (Ciclo del combustible nuclear)"

json_res, guion_res, img_prompt = visualmind(texto_a_procesar)

display(Markdown("## 💡 Ejecución de la Solución"))

display(Markdown("### Etapa 1: Destilación Lógica (JSON)"))
print(json.dumps(json_res, indent=4, ensure_ascii=False))

display(Markdown("### Etapa 2: Traducción Adaptativa (Feynman)"))
display(Markdown(guion_res))

display(Markdown("### Etapa 3: Refuerzo Visual (Prompt de Imagen)"))
display(Markdown(f"> **Prompt para DALL-E/Midjourney:** \n\n `{img_prompt}`"))
```
